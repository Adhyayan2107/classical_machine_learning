```python
import pandas as pd
pd.set_option('display.max_columns', 25)
df = pd.read_csv('telco_churn.csv')
```

```python
df.Churn.value_counts(normalize=True)
# Something which we want to predict is called positive class
```

```python
df.drop(columns=['customerID'] , inplace=True)
```

```python
columns = df.columns
YesNoColumns = []
for i in range(len(columns)):
    if(df[columns[i]].nunique() > 2):
        continue
    YesNoColumns.append(columns[i])
    print(df[columns[i]].value_counts(normalize=True))
```

```python
YesNoColumns
for col in YesNoColumns:
    df[col].replace({"Yes" : 1 , "No" : 0 , 1 : 1 , 0 : 0} , inplace=True)
```

```python
df[df.TotalCharges == " "].index
```

```python
df.drop(df[df.TotalCharges == " "].index , inplace=True)
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'])
```

```python
categorical_columns = df.select_dtypes('object')
categorical_columns
```

```python
dummies = pd.get_dummies(data=categorical_columns, drop_first=True , dtype=int)
```

```python
dummies
```

```python
df.drop(columns=categorical_columns.columns , inplace=True)
```

```python
df = pd.concat([df , dummies] , axis=1)
```

```python
df.columns
```

```python
X = df.drop(columns='Churn')
Y = df.Churn
from sklearn.model_selection import train_test_split
X_train  , X_test , Y_train , Y_test = train_test_split(X , Y , train_size=0.7 , random_state=100)
```

```python
X
```

```python
X_train
```

```python
numerical_columne = ['MonthlyCharges' , 'TotalCharges' , 'tenure']
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train[numerical_columne] = scaler.fit_transform(X_train[numerical_columne])
X_test[numerical_columne] = scaler.fit_transform(X_test[numerical_columne])
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train , Y_train)
```

```python
Y_pred = lr.predict(X_test)
```
